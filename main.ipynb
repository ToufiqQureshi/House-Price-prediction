```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
califor = fetch_california_housing()
```

```python
califor
```

```python
califor.keys()
```

```python
califor.feature_names
```

```python
#lets prepare the dataframe and these are my independent variables

df = pd.DataFrame(califor.data,columns=califor.feature_names)
```

```python
df.head()
```

```python
# Now add dependent variable
df['price'] = califor.target
```

```python
df.head()
```

```python
# understand the data 

df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
#lets visualize for understanding the data and see the correlation


df.corr()
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
#independent variables
X = df.iloc[:,:-1] #this means all columns except the last one

#dependent variable
y = df.iloc[:,-1] #this means last column
```

```python
# lets split the data into train and test

from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train , y_test = train_test_split(X,y,test_size=0.33,random_state=10) # i gave the test size as 0.33 which means 33% of the data will be used for testing
```

```python
# lets check the shape of the data

X_train.shape, y_train.shape
```

```python
# lets standardise the data

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() #initialize the scaler
```

```python
X_train=scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_test, X_train
```

```python
#Model Training 

from sklearn.linear_model import LinearRegression
```

```python
regresson = LinearRegression() #initialize the model
```

```python
regresson
```

```python
regresson.fit(X_train,y_train)
```

```python
regresson.coef_ #slope and intercept
```

```python
regresson.intercept_ #intercept
```

```python
#lets predict the output
y_pred = regresson.predict(X_test)
```

```python
y_pred
```

```python
#  Performance of the Model: Check how well the model is predicting

from sklearn.metrics import mean_squared_error  # Calculates the squared difference between actual & predicted values
from sklearn.metrics import mean_absolute_error  # Calculates the average absolute error (difference) between actual & predicted values
# Mean Absolute Error (MAE) - Shows the average error in predictions
print("Mean Absolute Error (MAE):", mean_absolute_error(y_test, y_pred))

# Mean Squared Error (MSE) - Penalizes larger errors more by squaring them
print("Mean Squared Error (MSE):", mean_squared_error(y_test, y_pred))

# Root Mean Squared Error (RMSE) - Square root of MSE, brings error back to original scale
print("Root Mean Squared Error (RMSE):", np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
# this is show us model accuracy 

from sklearn.metrics import r2_score
score = r2_score(y_test, y_pred)
print(score) #bad accuaracy
```

```python
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1) #actual accuracy also bad 
```
